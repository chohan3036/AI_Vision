### Python으로 커피머신 만들기
* 원두를 "Coffee Bean"이라 합니다. 원두를 넣으면 커피가 나오도록 문자열을 분리시키는 함수를 만들어 볼까요?

```
def coffee_machine(arg1):
  temp = arg1.split(' ')
  return temp[0]

input_value = "coffee bean"
cup = coffee_machine(input_value)
print(cup)
```

## Q1) 위의 결과 "coffee"가 나왔네요. 우리의 coffe_machine이 선택에 따라 뜨거운(hot), 차가운(ice) coffee를 만들어 낼 수 있도록 수정해 볼까요?
- function definition : def coffee_machine(arg1, arg2): <-- arg2 에는 hot과 ice의 선택을 할 수 있는 정수(1 또는 2)가 들어갑니다.
- Output : 'hot coffee' or 'ice coffee'

```
#Q1 Code
def coffe_machine(arg1, arg2):
  temp = arg1.split(' ')

  if (arg2 == 1):
    retTemp = 'hot ' + temp[0]
  
  else:
    retTemp = 'ice ' + temp[0]

  return retTemp

if __name__ == '__main__':
  input_value = 'coffee bean'
  cup = coffe_machine(input_value, 1)
  print(cup)
```

## Q2) 4칙연산(+, -, *, /)를 수행하는 각각의 함수를 만들고, 두개의 수를 input으로 받아 결과를 나타내는 프로그램을 구현하세요

```
def plus(int1, int2):
  return int1 + int2


def minus(int1, int2):
  return int1 - int2


def multiply(int1, int2):
  return int1 * int2


def divide(int1, int2):
  if (int2 != 0):
    return int1 // int2
  else:
    '나누기를 수행할 수 없습니다.'

if __name__ == '__main__':
  logs = []   # advanced : 사용자 누적 입력과 연산결과를 넣을 리스트

  while True:
    # 사용자 입력 받기
    print('1. 더하기    2. 빼기')
    print('3. 곱하기    4. 나누기')
    num1, num2, task = map(int, input('계산할 두 정수와 작업을 입력하세요 : ').split())

    result = 0
    if task == 1:
      result = plus(num1, num2)
    elif task == 2:
      result = minus(num1, num2)
    elif task == 3:
      result = multiply(num1, num2)
    else:
      result = divide(num1, num2)

    print('계산 결과 : ', result)
  
    logs.append((num1, num2, task, result))   # 결과를 리스트에 저장

    quit = input('종료하시려면 quit를 입력하세요 : ')
    if quit == 'quit':
      break

  print('사용자가 수행한 작업입니다.')
    for i in range(len(logs)):
        print('[', (i + 1), '번째 작업]  숫자1 : ', logs[i][0], ' 숫자2 : ', logs[i][1], ' 작업번호 : ', logs[i][2], ' 계산결과 : ', logs[i][3])
```

## List Comprehension

````
list_w = list(range(10))
print (list_w)
````

### [출력표현식 or conditon for 임시저장소 in iterable [if 조건식]]

````
list_e = [ x<5 for x in list_w]
print (list_e)
````

### 모든 리스트에 대한 AND, OR연산을 쉽게 해보자

````
list_r = [1, 2, 3, 4]

if (list_r[0]>2) and (list_r[1]>2) and (list_r[2] >2) and (list_r[3] > 2):
  print ("all of list_w is satisfied with 2>x condition")
else:
  print ("otherwise")
````

````
# AND 논리연산 --> all()
list_r = [1, 2, 3, 4]
if all( 2>x for x in list_r):
  print ("all of list_w is satisfied with 2>x condition")
else:
  print ("otherwise")
````

````
# OR 논리연산 --> any()
list_r = [1, 2, 3, 4]
if any( 2>x for x in list_r):
  print ("some items of list_w is satisfied with 2>x condition")
else:
  print ("otherwise")
````

##Q1) List Comprehension 을 이용하여 5(Row)x5(Column) Matrix를 만들고 다음과 같이 출력하는 프로그램을 구성하세요


```
['*', '*', '*', '*', '*']
['*', ' ', ' ', ' ', '*']
['*', ' ', ' ', ' ', '*']
['*', ' ', ' ', ' ', '*']
['*', '*', '*', '*', '*']
```

````
# Q1 Code
MAX_ = 5
mat = [[' ' for i in range(MAX_)] for j in range(MAX_)]
for i in range(MAX_):
  for j in range(MAX_):
     if (i == 0 or i == 4) or (j == 0 or j == 4):
       mat[i][j] = '*'
  print(mat[i])
````

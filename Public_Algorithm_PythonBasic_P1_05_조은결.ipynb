## 1. 1부터 N까지 짝수 출력하기
- N을 Standard Input에서 입력받으세요(Input() 이용)

````
#Q1 Code
n = int(input())
for i in range(1, n + 1):
    if i % 2 == 0:
        print(i)
````

## 2. 1부터 N까지 합 구하기
- N을 Standard Input에서 입력받으세요(Input() 이용)

````
#Q2 Code
n = int(input())
sum = 0
for i in range(1, n + 1):
    sum += i
print(sum)
````

###4.1 위의 MATRIX를 참고해서, 아래와 같이 출력하는 프로그램을 작성해 보세요


```
* * * * *
*       *
*       *
*       *
* * * * *
```



````
MAX_ROW = 5
MAX_COL = 5

MATRIX = [['*' for j in range(MAX_COL)] for i in range(MAX_ROW)]

for i in range(MAX_ROW):
    for j in range(MAX_COL):
        if i == 0 or i == 4:
            print(MATRIX[i][j], end = ' ')
        
        elif j == 0 or j == 4:
            print(MATRIX[i][j], end = ' ')
        
        else:
            print(' ', end = ' ')
    print()
    print()
````

#4.3 위의 MATRIX를 참고해서, 아래와 같이 출력하는 프로그램을 작성해 보세요


```
1               9
1 2           8 9
1 2 3       7 8 9 
1 2 3 4   6 7 8 9 
1 2 3 4 5 6 7 8 9
1 2 3 4   6 7 8 9
1 2 3       7 8 9
1 2           8 9
1               9
```   




````
mat = [[i for i in range(1, 10)] for _ in range(10)]

for i in range(10):
    for j in range(9):
        if i < 5:
            if j <= i or j >= 9 - i - 1:
                print(mat[i][j], end = ' ')
            else:
                print(' ', end = ' ')

        else:
            if j > i or j < 9 - i - 1:
                print(mat[i][j], end = ' ')
            else:
                print(' ', end = ' ')
    print()
````
